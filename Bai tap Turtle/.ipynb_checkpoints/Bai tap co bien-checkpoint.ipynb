```python
#Bai 1
import turtle
t=turtle.Turtle()
#chieu dai canh
a=200
#di chuyen toi diem ve 1
t.penup()
t.backward(a/2)
t.pendown()
#ve hinh thu vuong 1
t.left(90)
t.forward(a/2)
t.right(90)
t.forward(a)
t.right(90)
t.forward(a)
t.right(90)
t.forward(a)
t.right(90)
t.forward(a/2)
#di chuyen toi diem ve 2
t.penup()
t.right(90)
t.forward(a/2)
t.left(45)
t.forward(a/2)
t.pendown()
#ve hinh thu vuong 2
t.right(90)
t.forward(a/2)
t.right(90)
t.forward(a)
t.right(90)
t.forward(a)
t.right(90)
t.forward(a)
t.right(90)
t.forward(a/2)
turtle.done()
```
