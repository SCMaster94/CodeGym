```python
#Bai 1
import turtle
t=turtle.Turtle()
#di chuyen toi diem ve 1
t.penup()
t.backward(50)
t.pendown()
#ve hinh thu vuong 1
t.left(90)
t.forward(50)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(50)
#di chuyen toi diem ve 2
t.penup()
t.right(90)
t.forward(50)
t.left(45)
t.forward(50)
t.pendown()
#ve hinh thu vuong 2
t.right(90)
t.forward(50)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(50)
turtle.done()
```
