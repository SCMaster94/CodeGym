```python
#1
def tong_uoc_chan(num):
    tong_uoc_chan=0
    if num%2==0:
        for i in range (2,num+1,2):
            if num%i==0:
#                 print(i)
                tong_uoc_chan+=i
        else:
            return (tong_uoc_chan)
    else:
        print("So le thi lam gi co uoc chan")
while True:
    n=int(input('Nhap so nguyen duong n: '))
    if n>0:
        break
    else:
        print("Ban can nhap so nguyen duong")
total=tong_uoc_chan(n)
print(total)
```

```python
#2
def uoc_le_max(num):
    uoc_le_max=0
    for i in range (1,num+1,2):
        if num%i==0:
#             print(i)
            uoc_le_max=i
    else:
        return (uoc_le_max)
while True:
    n=int(input('Nhap so nguyen duong n: '))
    if n>0:
        break
    else:
        print("Ban can nhap so nguyen duong")
odd_max=uoc_le_max(n)
print(odd_max)
```

```python
#3
def so_lan_gap(paper_thick, thick):
    i=0
    while paper_thick<=thick:
        paper_thick=paper_thick*2
        i+=1
    return i
paper_thick=float(input("Nhap chieu day to giay can gap: "))
thick=float(input("Nhap chieu day can gap: "))
n=so_lan_gap(paper_thick, thick)
print(f"Can gap {n: .0f} lan de to giay day {paper_thick: .2f}mm co do day {thick: .2f}mm")
```

```python
#4
def check(num):
    i=1
    print_str=''
    for B in range (1,num):
        for A in range (1,B):
            if (A*A+B*B+1)%(A*B)==0:
#                 print(A*A+B*B+1,A*B)
                print_str+=f"Cap so {i} la: {A} va {B}\n"
                i+=1
    if print_str=='':
        print_str='Khong co cap so nao'
        return (print_str)
    else:
        return (print_str)
while True:
    n=int(input("Nhap co nguyen duong can check: "))
    if n>0:
        break
str_check=check(n)
print(str_check)
```
