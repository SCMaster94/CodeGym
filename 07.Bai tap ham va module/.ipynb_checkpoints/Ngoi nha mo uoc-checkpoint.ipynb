```python
#Ve ngoi nha
import turtle
def draw_rec(width, height, color, fill_color):  
    t.color(color)
    t.begin_fill()
    t.color(color)
    t.fillcolor(fill_color)
    t.pendown()
    for i in range(2):
        t.forward(width)
        t.left(90)
        t.forward(height)
        t.left(90)
    t.end_fill()
    t.penup()
def draw_isosceles_triangle(width, height, color, fill_color):  
    t.color(color)
    t.begin_fill()
    t.color(color)
    t.fillcolor(fill_color)
    top=t.pos()
    t.right(90)
    t.forward(height)
    midpoint=t.pos()
    t.left(90)
    t.forward(width/2)
    side=t.pos()
    t.pendown()
    t.begin_fill()
    t.backward(width)
    t.goto(top)
    t.goto(side)
    t.end_fill()
    t.penup()
    return(midpoint)
def draw_isosceles_trapezoid(width_1, width_2, height, color, fill_color):
    t.color(color)
    base=t.pos()
    t.fillcolor(fill_color)
    t.forward(width_1/2)
    top_right=t.pos()
    t.backward(width_1)
    top_left=t.pos()
    t.goto(base)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.forward(width_2/2)
    bot_left=t.pos()
    t.begin_fill()
    t.pendown()
    t.backward(width_2)
    t.goto(top_right)
    t.goto(top_left)
    t.goto(bot_left)
    t.end_fill()
    t.left(180)
    t.penup()
def draw_house(wall_w, wall_h, door_w, door_h, roof_w, roof_h, color, wall_color="white", door_color="white", roof_color="white"):
    base=t.pos()
    draw_rec(wall_w, wall_h, color, wall_color)
    t.forward((wall_w-door_w)/2)
    draw_rec(door_w, door_h, color, door_color)
#     t.forward(width*0.2)
#     t.left(90)
#     t.pendown()
#     t.forward(height*0.5)
    t.goto(base)
    t.left(90)
    t.forward(wall_h)
    t.right(90)
    base_roof=t.pos()
    t.forward(wall_w/2)
    #ve ong khoi
    draw_rec(15,50, color, wall_color)
    #ve khoi
    t.forward(2)
    t.left(90)
    t.penup()
    t.forward(60)
    t.pendown()
    t.forward(10)
    t.penup()
    t.backward(20)
    t.right(90)
    t.forward(10)
    t.left(90)
    t.forward(7)
    t.pendown()
    t.forward(15)
    t.penup()
    t.right(90)
    #ve mai nha
    t.goto(base_roof)
    t.forward(roof_w/2)
    t.left(90)
    t.forward(roof_h)
    t.right(90)
    draw_isosceles_triangle(roof_w,roof_h, color, roof_color)
def draw_tree(width, height, width_1, width_2, trunk_height,  color, color_leaf_1, color_leaf_2, color_leaf_3, trunk, sc):
    width=width*sc
    height=height*sc
    width_1=width_1*sc
    width_2=width_2*sc
    trunk_height=trunk_height*sc
    top=draw_isosceles_triangle(width, height, color, color_leaf_1)
    t.goto(top)
    width=width*1.2
    height=height*1.2
    top=draw_isosceles_triangle(width, height, color, color_leaf_2)
    t.goto(top)
    width=width*1.2
    height=height*1.2
    top=draw_isosceles_triangle(width, height, color, color_leaf_3)
    t.goto(top)
    draw_isosceles_trapezoid(width_1, width_2, trunk_height, 'blue', trunk)
t=turtle.Turtle()

#ve nha
color='#6489bd'
t.pensize(3)
wall_h=80
wall_w=200
door_h=50
door_w=40
roof_h=50
roof_w=150
roof_angle=40
ong_khoi_w=15
ong_khoi_h=40
d_khoi_ong=12
khoi_h=40
khoi_w=12
t.hideturtle()
#chuyen but
t.penup()
t.forward(-wall_w/2)
t.left(90)
t.forward(-wall_h/2)
t.right(90)
t.pendown()
draw_house(wall_w, wall_h, door_w, door_h, roof_w, roof_h, color, wall_color="white", door_color="white", roof_color="white")

#ve cay
t.pensize(1.5)
tan_cay_w=80
tan_cay_h=30
than_cay_w1=20
than_cay_w2=35
than_cay_h=60
tree_color='#7c9048'
leaf_color_1="#c6d4a1"
leaf_color_2="#7c9048"
leaf_color_3="#525f30"
trunk_color='#808080'
t.goto(200,100)
draw_tree(tan_cay_w, tan_cay_h, than_cay_w1, than_cay_w2, than_cay_h, tree_color, leaf_color_1, leaf_color_2, leaf_color_3, trunk_color,1)
t.goto(-140,30)
draw_tree(tan_cay_w, tan_cay_h, than_cay_w1, than_cay_w2, than_cay_h, tree_color, leaf_color_1, leaf_color_2, leaf_color_3, trunk_color,0.6)
turtle.done()
```
