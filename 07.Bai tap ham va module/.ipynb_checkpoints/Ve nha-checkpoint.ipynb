```python
#Ve nha
import turtle
t=turtle.Turtle()
t.pensize(2)
def draw_rec(width, height, color, fill_color):
    t.color(color)
    t.begin_fill()
    t.color(color)
    t.fillcolor(fill_color)
    for i in range(2):
        t.forward(width)
        t.left(90)
        t.forward(height)
        t.left(90)
    t.end_fill()
def draw_apar(width, height, color, wall_color, windown_color):
    t.pendown()
    draw_rec(width, height, color, wall_color)
    t.penup()
    t.forward(width/3*2)
    t.left(90)
    t.forward(height/3*2)
    t.left(90)
    t.pendown()
    draw_rec(width/3, height/3, color, windown_color)
    t.left(180)
    t.penup()
def draw_house(width, height, color, wall_color, windown_color, floor, n_apar):
    base_house=t.pos()        
    for i in range(floor):
        base_floor=t.pos()
        for i in range(n_apar):
                base_apar=t.pos()
                draw_apar(width, height, color, wall_color, windown_color)
                t.goto(base_apar)
                t.forward(width)
                base_apar=t.pos()        
        t.goto(base_floor)
        t.left(90)
        t.forward(height)
        t.right(90)
    t.goto(base_house)
def draw_gara(width, height, color, wall_color, door_color):
    t.pendown()
    draw_rec(width, height, color, wall_color)
    t.penup()
    t.forward(width*0.3)
    t.pendown()
    draw_rec(width*0.4, height*0.5, color, door_color)
    t.penup()
    t.forward(width*0.2)
    t.left(90)
    t.pendown()
    t.forward(height*0.5)
#Khai bao bien co dinh cho can nha    
width=60
height=100
color='blue'
wall_color='purple'
windown_color='white'
floor=2
n_apar=2

#can chinh vi chi can nha
t.penup()
t.goto(-width*n_apar/2,-height*floor/2)

#ve nha
draw_house(width,height,color, wall_color, windown_color, floor, n_apar)


#ve gara
t.forward(width*n_apar)
draw_gara(80, 90, 'blue', 'grey', 'green')

turtle.done()
```
