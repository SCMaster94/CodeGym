```python
#ve ngoi nha
import turtle
#Set backgroud collor
screen=turtle.Screen()
screen.bgcolor("Blue")
#Ve mai cam
h=turtle.Turtle()
h.fillcolor('orange')
h.begin_fill()
h.rt(45)
h.fd(100)
h.rt(100)
h.fd(100)
h.rt(80)
h.fd(100)
h.rt(100)
h.fd(100)
h.end_fill()
h.backward(100)
#Ve dau hoi
h.fillcolor('purple')
h.begin_fill()
h.rt(170)
h.fd(100)
h.lt(135)
h.fd(141)
h.end_fill()
#Ve tuong xanh
h.fillcolor('black')
h.begin_fill()
h.rt(90)
h.fd(141)
h.rt(90)
h.fd(141)
h.rt(90)
h.fd(141)
h.end_fill()

#Ve cua chinh
h.penup()
h.rt(90)
h.fd(141/3)
h.rt(90)
h.fd(141/2)
h.pendown()

h.fillcolor('green')
h.begin_fill()
h.lt(90)
h.fd(141/3)
h.rt(90)
h.fd(141/2)
h.rt(90)
h.fd(141/3)
h.rt(90)
h.fd(141/2)
h.end_fill()

h.penup()
h.home()
h.rt(45)
h.fd(100)
h.pendown()

h.fillcolor('yellow')
h.begin_fill()
h.rt(45)
h.fd(141)
h.rt(55)
h.fd(100)
h.rt(125)
h.fd(141)
h.end_fill()

h.penup()
h.backward(141)
h.fd(141/2)
h.rt(55)
h.fd(33)
h.pendown()

h.fillcolor('gray')
h.begin_fill()
h.fd(100/3)
h.lt(55)
h.fd(141/3)
h.lt(125)
h.fd(100/3)
h.lt(55)
h.fd(141/3)
h.end_fill()

turtle.done()
```
