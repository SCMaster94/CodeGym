```python
#ve ngoi nha
import turtle
import math
#Set backgroud collor
screen=turtle.Screen()
screen.bgcolor("Blue")
#set gia tri bien
canh_mai=100
canh_tuong=100*math.sqrt(2)
goc_nghieng=10

#Ve mai cam
h=turtle.Turtle()
h.fillcolor('orange')
h.begin_fill()
h.rt(45)
h.fd(canh_mai)
h.rt(canh_mai)
h.fd(canh_mai)
h.rt(80)
h.fd(canh_mai)
h.rt(canh_mai)
h.fd(canh_mai)
h.end_fill()
h.backward(canh_mai)
#Ve dau hoi
h.fillcolor('purple')
h.begin_fill()
h.rt(180-goc_nghieng)
h.fd(canh_mai)
h.lt(135)
h.fd(canh_tuong)
h.end_fill()
#Ve tuong xanh
h.fillcolor('black')
h.begin_fill()
h.rt(90)
h.fd(canh_tuong)
h.rt(90)
h.fd(canh_tuong)
h.rt(90)
h.fd(canh_tuong)
h.end_fill()

#Ve cua chinh
h.penup()
h.rt(90)
h.fd(canh_tuong/3)
h.rt(90)
h.fd(canh_tuong/2)
h.pendown()

h.fillcolor('green')
h.begin_fill()
h.lt(90)
h.fd(canh_tuong/3)
h.rt(90)
h.fd(canh_tuong/2)
h.rt(90)
h.fd(canh_tuong/3)
h.rt(90)
h.fd(canh_tuong/2)
h.end_fill()

h.penup()
h.home()
h.rt(45)
h.fd(canh_mai)
h.pendown()

h.fillcolor('yellow')
h.begin_fill()
h.rt(45)
h.fd(canh_tuong)
h.rt(55)
h.fd(canh_mai)
h.rt(125)
h.fd(canh_tuong)
h.end_fill()

h.penup()
h.backward(141)
h.fd(canh_tuong/2)
h.rt(55)
h.fd(canh_mai/3)
h.pendown()

h.fillcolor('gray')
h.begin_fill()
h.fd(canh_mai/3)
h.lt(55)
h.fd(canh_tuong/3)
h.lt(125)
h.fd(canh_mai/3)
h.lt(55)
h.fd(canh_tuong/3)
h.end_fill()

turtle.done()
```
